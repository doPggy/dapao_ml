```python
import tensorflow as tf
import numpy as np
```

```python
def add_layer(inputs, in_size, out_size, ac = None):
    Weights = tf.Variable(tf.random_normal([in_size, out_size]))
    biases  = tf.Variable(tf.zeros([in_size]) + 0.1)
    hypo    = tf.matmul(inputs, Weights) + biases
    if ac is None:
        outputs = hypo
    else:
        outputs = ac(hypo)
    return outputs
```

```python
x_data = np.linspace(-1, 1, 300)[:, np.newaxis].astype(np.float32)
noise  = np.random.normal(0, 0.05, x_data.shape).astype(np.float32)
y_data = (np.square(x_data) - 0.5 + noise).astype(np.float32)


xs = tf.placeholder(tf.float32, [None, 1])
ys = tf.placeholder(tf.float32, [None, 1])
```

```python
l1 = add_layer(xs, 1, 10, ac = tf.nn.relu)

predition = add_layer(l1, 10, 1, ac = None)

# loss = tf.reduce_mean(tf.reduce_sum(tf.square(ys - predition)),
#                      reduction_indices = [1])
loss = tf.reduce_mean(tf.reduce_sum(tf.square(ys - predition), axis = 1))
train_step = tf.train.GradientDescentOptimizer(0.1).minimize(loss)


init      = tf.global_variables_initializer()
```

```python
with tf.Session() as s:
    s.run(init)
    for i in range(1000):
        s.run(train_step, feed_dict = {xs:x_data, ys:y_data})
        if i % 50:
            print(s.run(loss, feed_dict = { xs:x_data, ys:y_data}))
```
